```python
import numpy as np
from astropy.modeling import models
import astropy.units as u
from specutils import Spectrum1D, SpectralRegion
import matplotlib.pyplot as plt
```

```python
datafile = '20200428_scan2_unblocked_data.csv'

k = np.genfromtxt(datafile, delimiter=',')[1:800,1]
a = np.genfromtxt(datafile, delimiter=',')[1:800,2]

plt.plot(k, a)
plt.show()
```

```python
spectrum = Spectrum1D(flux=a*u.dimensionless_unscaled, spectral_axis=k/u.cm)

plt.plot(spectrum.spectral_axis, spectrum.flux) 
plt.xlabel('Wavenumber ({})'.format(spectrum.spectral_axis.unit)) 
plt.ylabel('Absorption') 
plt.grid(True) 
```

```python
noise_region = SpectralRegion((.528+1.7323e4)*u.um, (.532+1.7323e4)*u.um)

from specutils.fitting import find_lines_derivative
lines = find_lines_derivative(spectrum,flux_threshold=0.15)
lines
```
