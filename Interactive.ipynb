# Interactive Peak Fitting

## Running find_peaks will identify possible peak locations to be used as the initial parameters for a least-squares fit. 
The a-peaks, indicated by solid lines, are locations where the code is almost certain there is a peak (via first-derivative analysis). The b-peaks, indicated by dashed lines, are locations where the code is less certain there is a peak but has still determined a peak to possibly be present (via second-derivative analysis).

```python
%run find_peaks.ipynb
```

## If the user thinks a peak has been missed, or a region has been falsely identified as a peak, they can add or remove peaks before the spectrum is fit using edit_peaks.

If the user does not think this is necessary, they can ignore the next cell and continue to fitting. Zooming in on the above plot by double-clicking it and ensuring all peaks look accounted for is recommended.

```python
%run edit_peaks.ipynb
```

## Once the possible peak locations to be used as initial parameters are to the user's satisfaction, the user can perform a least-squares fit by running fit_peaks.

This uses Geoffrey Lentner's dataphile package to generate a GUI which allows the user to further adjust the initial parameter guesses before fitting if desired. To select a peak, click one of the ovals to the left of the sliders (you may need to enlarge the plot by dragging its corner out in order to do this). Adjust the initial parameters for each Gaussian using the sliders, then click the "fit" button when ready to perform the least-squares fit.

```python
%run fit_peaks.ipynb
```

## After the fit is performed, one can generate a table listing the best-fit parameter values and their uncertanties as well as plot the peak locations using plot_peaks.

If you are unsatisfied with the fit, I recommend either adding or removing peaks by running edit_peaks again or changing the initial fit parameters by running fit_peaks again and adjusting the sliders. So far I've found that the code does a pretty good job of finding the peaks and fitting them on its own.

```python
%run plot_peaks.ipynb
```
