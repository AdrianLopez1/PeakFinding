# Interactive Peak Fitting

## Running find_peaks will identify possible peak locations to be used as the initial parameters for a least-squares fit. 
For the data file, a .csv file must be supplied with the wavenumbers in the first column and the absorption values in the second column.

The a-peaks, indicated by solid lines, are locations where the code is more certain there is a peak (via first-derivative analysis). The b-peaks, indicated by dashed lines, are locations where the code is less certain there is a peak but has still determined a peak to possibly be present (via second-derivative analysis).

```python
datafile = "bend_scan_44p8_45p8_modified"          # Name of data file
threshold = 10                                     # Minimum peak height as percentage of maximum absorption
partition = 2                                      # Which partition of the data to use
%run find_peaks.ipynb
```

## If the user thinks a peak has been missed, or a region has been falsely identified as a peak, they can add or remove peaks before the spectrum is fit using edit_peaks.

If the user does not think this is necessary, they can ignore the next cell and continue to fitting. Zooming in on the above plot by double-clicking it and ensuring all peaks look accounted for is recommended.

If there are many small fluctuations misidentified as noise, or many large peaks missed, the user can try adjusting the threshold variable in find_peaks and running find_peaks again. This sets the minimum amplitude of what will be identified as a peak.

To add a peak, enter the wavenumber of the peak to be added in the left text box and click "Add peak." To remove a peak, enter the index of the peak to be removed (displayed below the vertical line marking the peak) and click "Remove peak." Removing peaks will update the indices of other peaks, so use the updated plot below if you want to remove several peaks.

## Once the possible peak locations to be used as initial parameters are to the user's satisfaction, the user can perform a least-squares fit by running fit_peaks.

This uses Geoffrey Lentner's dataphile package to generate a GUI which allows the user to further adjust the initial parameter guesses before fitting if desired. To select a peak, click one of the ovals to the left of the sliders (you may need to enlarge the plot by dragging its corner out in order to do this). Adjust the initial parameters for each Gaussian using the sliders or text boxes, then click the "fit" button when ready to perform the least-squares fit.

If the cell below appears blank after being run, try running it again. If the problem persists, try restarting the kernel.
